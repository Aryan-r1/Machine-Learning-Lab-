```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Mail.csv')
```

```python
print(df)
```

```python
data = df.where((pd.notnull), '')
```

```python
data.head()
```

```python
data.head()
```

```python
data.shape
```

```python
data.loc[data['Category'] == 'spam', 'Category'] = 0
data.loc[data['Category'] == 'ham', 'Category'] = 1
```

```python
X = data['Message']

Y =data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_test_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

```python
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy Training Data : ', accuracy_on_training_data)
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy Test Data : ', accuracy_on_test_data)
```

```python
input_your_mail = ["I never realized that you were so embarassed by your accomodations. I thought you liked it, since i was doing the best i could and you always seemed so happy about ""the cave"". I'm sorry I didn't and don't have more to give. I'm sorry i offered. I'm sorry your room was so embarassing."]

input_data_features = feature_extraction.transform(input_your_mail)

prediction = model.predict(input_data_features)

print(prediction)

if(prediction[0] == 1):
    print('Ham mail')
else:
    print('Spam Email')
```
